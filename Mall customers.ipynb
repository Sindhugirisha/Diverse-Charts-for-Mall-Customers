# BAR and BARh Chart

```python
df[["Age","Spending Score (1-100)"]].plot(kind='hist',bins=10)
```

```python
df[["Age","Spending Score (1-100)"]].plot(kind='hist',bins=10,alpha=0.5)
```

```python
## Group by
Rank_totals = round(df.groupby("Gender")["Spending Score (1-100)"].mean(),2)
print(Rank_totals)
Rank_totals.plot(kind='bar')
```

```python
Rank_totals=df.groupby(["CustomerID"])[["Spending Score (1-100)","Annual Income (k$)"]].mean()
#print rank total
Rank_totals.plot(kind='bar',alpha=0.6,stacked=True)
```

# PIE CHART

```python
Rank_totals.plot(kind='pie',subplots=True)
```

```python
Rank_totals.plot.bar(color=['green','red'],subplots =True,rot=45)
```

```python
df['Annual Income (k$)'].plot(kind='box')
```

```python
import numpy as np

q2 = np.quantile(df['Annual Income (k$)'],0.5)
print('q2',q2)
q1 = np.quantile(df['Annual Income (k$)'],0.25)
print('q1',q1)
q3 = np.quantile(df['Annual Income (k$)'],0.75)
print('q3',q3)

iqr=(q3-q1)
print('iqr',iqr)

[q3-1.5*iqr,q1+1.5*iqr]

min,max=q3-1.5*iqr,q1+1.5*iqr
iqrrange=[min,max]
iqrrange
```

```python
annual_income = df[(df['Annual Income (k$)'] >= min) & (df['Annual Income (k$)'] <= max)]
annual_income.count()
```

```python
annual_income['Annual Income (k$)'].plot(kind='box')
```
